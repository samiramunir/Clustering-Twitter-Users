```python
import pandas as pd 
import numpy as np
import os
os.listdir('data_collection/data')
```

```python
df = pd.read_csv('data_collection/data/unique_users.csv')
df.head()
```

```python
# Only getting the summary for the retweet_count and total_weight column
df.drop(['Unnamed: 0', 'user_id'], axis =1).describe()
```

#### The number of retweets for each NYT posts 

From the graph below we can see that 

```python
import plotly as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)

data = [go.Histogram(x=df.retweet_count, 
                     marker=dict(color='rgb(8, 81, 156)'), 
                     nbinsx = 100, opacity = .50)]

layout = go.Layout(
    title='Retweet Count vs Number of Aricles ',
    xaxis=dict(
        title='Number of Retweets',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Number of NYT Tweets',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

py.offline.iplot(go.Figure(data=data, layout=layout), filename='basic histogram')
```
